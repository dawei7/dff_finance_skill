```python

import plotly.express as px
import yfinance as yf


ticker_symbol = "MSFT"

ticker = yf.Ticker(ticker_symbol)

# get stock info

df = ticker.history(period="max")

long_name = ticker.info["longName"]

fig = px.line(df, x=df.index, y="Close", title=f'{long_name} ({ticker_symbol}) - Share price plot in USD')
fig.show()


df
```
